```python
import requests, zipfile, io

url = "https://github.com/abdumaa/HiQualProp/archive/refs/heads/main.zip"
r = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall("data/raw/hqp")

import os
for root, dirs, files in os.walk("data/raw/hqp"):
    for f in files:
        print(os.path.join(root, f))
```
